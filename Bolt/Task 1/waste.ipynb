```python
visualize_weekly_numbers(
    df_products[
        (df_products['date'] >= '2023-07-03')
        & df_products['category_level_1'].isin(
            [
                'Meat  fish', 
                'Dairy  eggs',
                'Bakery',
                'Fresh  ready',
                'Cheese'
            ]
        )
    ],
    'category_level_1',
    show_by="absolute",
    metrics=['waste_num_of_items'],
    y_label='Waste items'
)
```

For some categories we have a significant spike in wasted items. Let's check other calendar dependencies for them.

```python
(
    df_products
    .groupby('category_level_1')
    .agg(
        wasted_items=('waste_num_of_items', 'sum'),
        wasted_cost=('waste_cogs', 'sum')
    )
    .sort_values('wasted_cost', ascending=False)
)
```

```python
top_waste_categories = list(
    df_products
    .groupby('category_level_1')
    .agg(
        wasted_items=('waste_num_of_items', 'sum'),
        wasted_cost=('waste_cogs', 'sum')
    )
    .sort_values('wasted_cost', ascending=False)
    .iloc[:10]
    .index
)
top_waste_categories
```

```python
# Weekday correlation
(
    df_products
    [df_products['category_level_1'].isin(top_waste_categories)]
    .groupby('weekday')
    .agg(waste_num_of_items=('waste_num_of_items', 'sum'))
    .sort_index()
    .plot(kind='bar', y='waste_num_of_items')
)
```

Ok, then the waste is more or less equal during the week. Checking if it depends on whether it's the end or beginning of the month.

```python
filtered_df = df_products[df_products['category_level_1'].isin(top_waste_categories)]

daily_waste = filtered_df.groupby('day_of_month').agg(waste_num_of_items=('waste_num_of_items', 'sum')).sort_index()
daily_waste['cumulative_moving_average'] = daily_waste['waste_num_of_items'].cumsum() / range(1, len(daily_waste) + 1)

plt.figure(figsize=(12, 6))
plt.plot(daily_waste.index, daily_waste['waste_num_of_items'], label='Actual Data', marker='o')
plt.plot(daily_waste.index, daily_waste['cumulative_moving_average'], label='Cumulative Moving Average', color='green')

plt.legend()
plt.title('Number of Discarded Items by Day of Month')
plt.xlabel('Day of Month')
plt.ylabel('Number of Items')

plt.show()
```

Ok, there is a slight trend, but the incline and amount of data doesn't allow to state that it is not a random dependency.
