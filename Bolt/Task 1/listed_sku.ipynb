# Research

How number of unique listed SKUs correlates with number of unique SKU ordered.

## Overall correlation

```python
(
    df_products
    .groupby('category_level_2')
    .agg(
        {
            'unique_skus_listed':'median',
            'unique_skus_sold':'median'
        }
    )
)
```

```python
# I assume that the same SKU can not be listed in two different categories
df_grouped = df_products.groupby('date').agg({
    'unique_skus_listed': 'sum',
    'unique_skus_sold': 'sum'
}).resample('W').median().sort_index()

# Calculating Pearson correlation
correlation = df_grouped.corr().iloc[0, 1]
print(f"Pearson correlation coefficient: {correlation}")

# Plotting the time series data
fig, ax1 = plt.subplots(figsize=(12,6))

color = palette[0]
ax1.set_xlabel('Weeks')
ax1.set_ylabel('Unique SKUs Listed', color=color)
ax1.plot(df_grouped.index, df_grouped['unique_skus_listed'], color=color, label='Unique SKUs Listed')
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = palette[1]
ax2.set_ylabel('Unique SKUs Sold', color=color)
ax2.plot(df_grouped.index, df_grouped['unique_skus_sold'], color=color, label='Unique SKUs Sold')
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout() 
fig.autofmt_xdate()


# Plotting histograms for distribution of SKUs listed and sold per category
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

df_median_distribution_per_cat_2 = (
    df_products
    .groupby('category_level_2')
    .agg(
        {
            'unique_skus_listed':'median',
            'unique_skus_sold':'median'
        }
    )
)

sns.histplot(df_median_distribution_per_cat_2['unique_skus_listed'], kde=True, color=palette[0], ax=axs[0])
axs[0].set_title('Distribution of Unique SKUs Listed')
axs[0].set_xlabel('Median number of unique SKUs listed per category_2')
axs[0].set_ylabel('Number of categories_2')

sns.histplot(df_median_distribution_per_cat_2['unique_skus_sold'], kde=True, color=palette[1], ax=axs[1])
axs[1].set_title('Distribution of Unique SKUs Sold')
axs[1].set_xlabel('Median number of unique SKU sold per category_2')
axs[1].set_ylabel('Number of categories_2')

plt.tight_layout()
plt.show()
```

## By categories
Now, I would like to examine how the relationship between the number of SKUs listed and the number of SKUs sold varies across each category. 

I will sort the categories by their median number of SKUs listed and visually inspect the results using a chart.

```python
df_products.columns
```

```python
from sklearn.preprocessing import MinMaxScaler

# Initialize a MinMaxScaler
scaler = MinMaxScaler()

for x, y in (
    df_products
    .groupby(['category_level_1', 'date'])
    .agg({'unique_skus_listed': 'sum'})
    .reset_index()
    .groupby('category_level_1')
    .agg({'unique_skus_listed': 'median'})
    .sort_values(by='unique_skus_listed', ascending=False)
    .reset_index()
    .values
):
    print(f'Category {x}, {y} median SKUs listed:')
    df_grouped = (
        df_products
        [df_products['category_level_1'] == x]
        .groupby('date')
        .agg(
            {
                'unique_skus_listed': 'sum',
                'unique_skus_sold': 'sum',
                'revenue_after_discount': 'sum'
            }
        )
        .resample('W')
        .median()
        .sort_index()
    )
    
    # Normalize the data
    df_grouped_normalized = pd.DataFrame(
        scaler.fit_transform(df_grouped),
        columns=df_grouped.columns,
        index=df_grouped.index
    )
    
    fig, ax = plt.subplots(figsize=(12, 6))
    
    # Plotting all normalized metrics on the same axis
    ax.plot(df_grouped_normalized.index, df_grouped_normalized['unique_skus_listed'], color=palette[0], label='Unique SKUs Listed')
    ax.plot(df_grouped_normalized.index, df_grouped_normalized['unique_skus_sold'], color=palette[1], label='Unique SKUs Sold')
    ax.plot(df_grouped_normalized.index, df_grouped_normalized['revenue_after_discount'], color=palette[3], label='Revenue After Discount')
    
    ax.set_xlabel('Weeks')
    ax.set_ylabel('Normalized Values')
    ax.legend()
    
    fig.tight_layout()
    fig.autofmt_xdate()
    plt.show()
```

# Median shares per category_2
As we can see from the above results, there is significant variation from one category to another. Let's examine the data at a more detailed level of granularity.

```python
df = (
    df_products
    .groupby('category_level_2')
    .agg(
        {
            'unique_skus_listed': 'median',
            'unique_skus_sold': 'median'
        }
    )
)

df['share'] = round(df['unique_skus_sold'] / df['unique_skus_listed'] * 100).astype(str) + '%'

df[df['unique_skus_listed']>=50].sort_values('unique_skus_listed', ascending=False).fillna(0)
```

Most categories aren't overloaded, so user choice isn't too dispersed across SKUs. However, the dynamics of unique SKUs sold, while positively correlated, don't always match the listing dynamics. The Pearson correlation coefficient is 0.48. 

This means we have categories with many SKUs but no user demand, cluttering the product list, potentially confusing users, and overly focusing their attention on top SKUs. This could increase bounce rates and impact revenue.

Proposed solutions include:

‚Äî Investigating these cases further and experimenting with algorithms to adjust sorting/grouping in overloaded categories.

‚Äî Modifying the UI to organize and represent products in these categories more clearly for the user.
