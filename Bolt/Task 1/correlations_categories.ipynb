```python
grouped_df = df_products.groupby(['date', 'category_level_0', 'category_level_1']).sum().reset_index()
pivot_df = grouped_df.pivot_table(index=['date', 'category_level_0'], columns='category_level_1', values='num_of_sold_sku_items')

# Empty DataFrame to store results
correlation_results = pd.DataFrame(columns=['category_level_0', 'category_level_1_x', 'category_level_1_y', 'correlation'])

# Calculate correlations within each category_level_0
for category_level_0, group in pivot_df.groupby(level=1):  # level=1 corresponds to 'category_level_0'
    # We need to drop the 'category_level_0' level to ensure the correlation calculation works as expected
    group = group.droplevel('category_level_0')
    correlations = group.corr().stack()
    correlations.index.names = ['category_level_1_x', 'category_level_1_y']
    correlations = correlations.reset_index(name='correlation')
    # Filter out self-correlations and ensure unique pairs
    correlations = correlations[
        (correlations['category_level_1_x'] < correlations['category_level_1_y'])
    ]
    # Add the category_level_0 for reference
    correlations['category_level_0'] = category_level_0
    # Append to the results DataFrame
    correlation_results = pd.concat([correlation_results, correlations], ignore_index=True)

# Filtering correlations to check further the most positive correlated
filtered_correlations = correlation_results[correlation_results['correlation'] > 0.5]

print(filtered_correlations)
```

```python
df_products = df_products.sort_values('date')

for cat_0, cat_1_x, cat_1_y, corr_score in filtered_correlations.values:
    print(f"Category 1: {cat_1_x}, Category 2: {cat_1_y}, Correlation: {corr_score}")
    filtered_df = df_products[df_products['category_level_1'].isin([cat_1_x, cat_1_y])]
    
    # Calculating agg values for number of sold SKUs per category
    aggregated_data = filtered_df.groupby(['date', 'category_level_1']).agg({'num_of_sold_sku_items':'sum'}).reset_index()
    aggregated_data.set_index('date', inplace=True)
    
    # Plotting
    fig, ax1 = plt.subplots(figsize=(12,6))
    ax2 = ax1.twinx()  # Create a second y-axis
    axes = [ax1, ax2]
    
    for i, category in enumerate([cat_1_x, cat_1_y]):
        category_data = aggregated_data[aggregated_data['category_level_1'] == category].copy()
        
        # Calculate the 7-day rolling average
        category_data['7_day_avg'] = category_data['num_of_sold_sku_items'].rolling(window=7).mean()
        
        # Plot the original data and 7-day rolling average on appropriate axis
        ax = axes[i]
        ax.plot(
            category_data.index, 
            category_data['num_of_sold_sku_items'], 
            label=f"{category} Sales", 
            alpha=0.1, 
            color=palette[i]
        )
        ax.plot(
            category_data.index, 
            category_data['7_day_avg'], 
            label=f"{category} 7-Day Avg", 
            linestyle='--', 
            color=palette[i]
        )
        
        # Set y-axis label for each category
        ax.set_ylabel(f'Number of Sold SKU Items for {category}', color=palette[i])
        
        # Ensure the tick params are set for both axes
        ax.tick_params(axis='y', labelcolor=palette[i])
    
    # General plot settings
    ax1.set_xlabel('Date')
    fig.suptitle('Sales Trends with 7-Day Rolling Average')
    
    # Handling legend: Collecting all labels and handles and then display them together
    handles, labels = [], []
    for ax in axes:
        for handle, label in zip(*ax.get_legend_handles_labels()):
            handles.append(handle)
            labels.append(label)
    fig.legend(handles, labels, loc='upper left')
    
    plt.xticks(rotation=45)
    fig.autofmt_xdate()
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])  # Adjusting the layout to make room for the title
    plt.show()
```
